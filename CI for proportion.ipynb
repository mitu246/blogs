```python
def get_prop(mu, sigma, sample_size):
    return np.mean(np.where(np.random.normal(mu, sigma, sample_size) > 0, 1, 0))
```

```python
# Simulation of growth:
n_days = 50
growth = 0.008
start_rev = 4
start_var = 20
n = 1000
prop_ls = []
for i in range(n_days):
    prop_ls.append(get_prop(start_rev, start_var, n))
    start_rev*=1+growth
    start_var*=1-growth
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (12,5))
sns.lineplot(x = [f"day_{day}" for day in np.arange(1,n_days+1)], y = prop_ls)
plt.title("Metric (% of total shoppers with atleast one product bought) Growth")
plt.xticks(rotation = 60)
plt.tight_layout()
```
